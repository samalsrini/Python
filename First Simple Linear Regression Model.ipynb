```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mat
```

```python
ds = pd.read_csv('Desktop/Python Codes/Simple.csv')
```

```python
ds
```

```python

x= ds.drop(['AvgUnempRate'],axis=1) # or write this incase of one independent variable x = ds.RGDPGrowth
y = ds.AvgUnempRate
print(x.head(10))
print("\n")  # for a space in the output
print(y.head(10))
```

```python
# splitting the data into training and testing 
# sklearn is the package/library for ML functions

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (x,y, test_size=0.3, random_state=6)

print(x_train)
print("\n")
print(y_train)
```

```python
print(x_test)
print("\n")
print(y_test)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)   # fitting my training data to algo for understadning pattern
```

```python
M=regressor.coef_                # slope
C=regressor.intercept_           # interecept
```

```python
print("Coefficent(M)=", regressor.coef_)
print("Intercept (C)=", regressor.intercept_)
```

```python
# predicting the test set results
y_predict = regressor.predict(x_test)
print(y_predict)
print("\n")
print(y_test)
```

```python
# understanding error. Lower the error, that good is to predict

from sklearn import metrics
print ("MAE=", metrics.mean_absolute_error(y_test, y_predict))
print ("MSE=", metrics.mean_squared_error(y_test, y_predict))
print ("RMSE", np.sqrt(metrics.mean_squared_error(y_test, y_predict)))
```

```python
from sklearn.metrics import r2_score  # how good is the model based on R2--correltaion coefficient..R2 higher the better

print('R2:', r2_score(y_test, y_predict))
```

```python
# only around 20% of the model is accurate/20% of the variance explained/20% confident.R2 ranges from -1 to +1
```
